```python
from dotenv import find_dotenv, load_dotenv
from transformers import pipeline
from transformers import VitsModel, AutoTokenizer
from langchain import PromptTemplate, LLMChain, OpenAI
import os
import requests
import torch
from IPython.display import Audio
```

```python
load_dotenv(find_dotenv())
HUGGINGFACEHUB_API_TOKEN = os.getenv('HUGGINGFACEHUB_API_TOKEN')
headers = {"Authorization": f"Bearer {HUGGINGFACEHUB_API_TOKEN}"}
```

```python
def img2text(url):
    image_to_text = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large")

    text = image_to_text(url)[0]['generated_text']
    print(text)
    return text
```

```python
def generate_story(scenario):
    template = """
    You are story teller;
    You can generate  a short story based on a simple narrative, the story should be no more than 30 words.
    CONTEXT: {scenario}
    STORY:
    """
    API_URL = "https://api-inference.huggingface.co/models/google/gemma-7b"
    payloads = {
        "inputs": scenario
    }
    response = requests.post(API_URL, headers=headers, json=payloads)
    return response.json()[0]['generated_text']
```

```python
winter = '/home/mydum/Desktop/ML/pracice/ ðŸ¤—HiggingFace/AudioImage/winter.jpg'
scenario = img2text(winter)
```

```python
story = generate_story(scenario)
story
```

```python
def text2speech(message):
    API_URL = "https://api-inference.huggingface.co/models/facebook/mms-tts-eng"
    payloads = {
        "inputs": message
    }
    response = requests.post(API_URL, headers=headers, json=payloads)
    with open('audio.flac', 'wb') as f:
        f.write(response.content)
```
