```python
mr_ms = df[df.Sex == "unknown"].Name.apply(lambda x: pd.Series(x.split("; ")[1].split()[0]))
mr_ms[0].unique()
```

```python
sex_map = {"Mr.": "male",    
           "Miss.": "female",
           "Mrs.": "male",
           "Master.": "male"}
```

```python
df.loc[df.Sex == 'unknown', 'Sex'] = df[df.Sex == 'unknown'].Name.apply(
    lambda x: sex_map.get(x.split('; ')[1].split()[0], 'unknown')
)
```

```python
data = df
data["is_male"] = df["Sex"] == "male"
data.head()
```

```python
sns.barplot(x="is_male", y="Survived", data=data, palette="winter")
plt.title("Sex - Survived")
plt.show()
```

```python
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=data, palette="winter")
plt.title("Sex - Survived")
plt.show()
```

```python
sns.catplot(x="Sex", y="Age", data=data, kind="box")
```

```python
sns.catplot(x="Sex", y="Age", hue="Pclass", data=data, kind="box")
```
